```python
import sklearn.datasets
import numpy as np
import pandas as pd

iris = sklearn.datasets.load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
stds = df.std()
print(stds)
```

```python
means = df.mean()
print(means)
```

```python
maxs = df.max()
print("\nmax")
print(maxs)
mins = df.min()
print("\nmin")
print(mins)
```

```python
d = df.describe()
print(d)
```
