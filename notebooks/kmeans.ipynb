```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# random state is fixed to experiment with the differences more easily

X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  random_state=0,
                 )
kmeans = KMeans(n_clusters=4,
                init='k-means++',
                random_state=0,
               )
kmeans.fit(X)
```

```python
# get some data out for visualizing

centers = kmeans.cluster_centers_
labels = kmeans.labels_

# print the labels
print(set(labels))
```

```python
import numpy as np

# get the labels
# numpy used to enable fancy slicing
label_0 = np.array([X[i] for i in range(len(labels)) if labels[i] == 0])
label_1 = np.array([X[i] for i in range(len(labels)) if labels[i] == 1])
label_2 = np.array([X[i] for i in range(len(labels)) if labels[i] == 2])
label_3 = np.array([X[i] for i in range(len(labels)) if labels[i] == 3])
```

```python
from matplotlib import pyplot as plt

plt.xlabel("feature 1")
plt.ylabel("feature 2")

# add the clusters
plt.scatter(label_0[:, 0], label_0[:, 1])
plt.scatter(label_1[:, 0], label_1[:, 1])
plt.scatter(label_2[:, 0], label_2[:, 1])
plt.scatter(label_3[:, 0], label_3[:, 1])

# plot the cluster centers
centers_X = [point[0] for point in centers]
centers_Y = [point[1] for point in centers]
plt.scatter(centers_X, centers_Y, c='#000000', marker='X')
```
