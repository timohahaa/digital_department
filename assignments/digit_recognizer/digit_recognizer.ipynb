```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
file_train = pd.read_csv('train.csv')

df_train = pd.DataFrame(file_train)
df_train.head()
```

```python
file_test = pd.read_csv('test.csv')

df_test = pd.DataFrame(file_test)
```

```python
labels = df_train.iloc[:, 0]
pixels = df_train.iloc[:, 1:]
```

```python
from sklearn import svm

svc = svm.SVC(
    kernel='poly',
    C=1,
    degree=2,
)

svc.fit(pixels, labels)

predicted = svc.predict(df_test)

```

```python
ans = pd.DataFrame({'ImageId': range(1,len(predicted) + 1), 'Label': predicted})
ans.head()
```

```python
ans.to_csv('ans.csv', index=False)
```
