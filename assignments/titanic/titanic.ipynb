```python
import pandas as pd
import numpy as np
import matplotlib as plt
#from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('train.csv')
df['Sex'] = df['Sex'].map({'male': 0,'female': 1})

df.head()
```

```python
labels = df.iloc[:, 1]

#df.columns[df.isnull().any()].tolist()
#['Age', 'Cabin', 'Embarked']

df['Age'].fillna(df['Age'].mean(), inplace=True)
df = df.loc[:, ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]

df.head()
```

```python
scaler = StandardScaler()

df[['Age']] = scaler.fit_transform(df[['Age']])
df[['Fare']] = scaler.fit_transform(df[['Fare']])

df.head()
```

```python
test_data = pd.read_csv('test.csv')
test_data['Sex'] = test_data['Sex'].map({'male': 0,'female': 1})

#test_data.columns[test_data.isnull().any()].tolist()
#['Age', 'Fare', 'Cabin']

test_data['Age'].fillna(test_data['Age'].mean(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].mean(), inplace=True)

test_data = test_data.loc[:, ['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
test_data[['Age']] = scaler.fit_transform(test_data[['Age']])
test_data[['Fare']] = scaler.fit_transform(test_data[['Fare']])

```

```python
from sklearn import svm

svc = svm.SVC(
    kernel='poly',
    C=1,
    degree=2,
)

svc.fit(df, labels)
```

```python
predicted = svc.predict(test_data.iloc[:, 1:])
```

```python
ans = pd.DataFrame({'PassengerId': test_data.loc[:, 'PassengerId'], 'Survived': predicted})
ans.head()
```

```python
ans.to_csv('ans.csv', index=False)
```
